# Pixi Package Manager Test - Install Libraries

Test installing libraries via Pixi package manager (both PyPI and conda).

```python
# Install pytest from PyPI via pixi
!pixi add --pypi pytest
```

```python
import pytest
```

```python
# Verify pytest is functional
version = pytest.__version__
print(f"pytest version: {version}")
assert version, "pytest version should not be empty"
print("✓ pytest installed and functional")
```

```python
# Install conda-build from conda (not available on PyPI)
!pixi add conda-build
```

```python
import conda_build
```

```python
# Verify conda-build is functional
version = conda_build.__version__
print(f"conda-build version: {version}")
assert version, "conda-build version should not be empty"
print("✓ conda-build installed and functional")
```
